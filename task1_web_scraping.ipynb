```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = "http://quotes.toscrape.com/page/{}/"
quotes_data = []

for page in range(1, 6):
    print(f"Scraping Page {page}...")

    url = base_url.format(page)

    try:
        response = requests.get(url)
        response.raise_for_status()   # Check if request successful
    except requests.exceptions.RequestException as e:
        print("Error accessing website:", e)
        break

    soup = BeautifulSoup(response.text, "html.parser")
    quotes = soup.find_all("div", class_="quote")

    if not quotes:
        print("No quotes found on this page.")

    for quote in quotes:
        text = quote.find("span", class_="text").text
        author = quote.find("small", class_="author").text
        tags = [tag.text for tag in quote.find_all("a", class_="tag")]

        quotes_data.append({
            "Quote": text,
            "Author": author,
            "Tags": ", ".join(tags)
        })

if quotes_data:
    df = pd.DataFrame(quotes_data)
    df.to_csv("quotes_multi_page_dataset.csv", index=False)

    print("\nData Scraped Successfully!")
    print("Total Quotes Scraped:", len(df))
    print(df.head())
else:
    print("No data scraped. Please check internet connection.")
```
